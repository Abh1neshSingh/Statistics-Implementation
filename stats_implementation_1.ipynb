```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings("ignore")
```

```python
# normal distribution / gaussian distribution  >> majority of datapoint  is clustered around mean
mu, sigma=0,0.1
s=np.random.normal(mu, sigma, size=10000)
print("count of data point ",len(s))
s
```

```python
# above data follow in normal distributions 
```

```python
# now show the plot
plt.hist(s,30, density= True)
plt.show()
```

```python
sns.distplot(s)
```

```python
# now uniform distribution -- all the outcome equal probability to occur in range
```

```python
u=np.random.uniform(1,0,1000)
```

```python
plt.hist(u)
plt.show()
```

```python
# log normal distn -- whose logarithm is normally distributed -- right skewed 
```

```python
l=np.random.lognormal(mean=0, sigma=0.5, size=10000)
l=l[np.where(abs(l)<5)]
plt.hist(l,bins=100)
plt.show()
```

```python
# now convert log normal distribution  to normal distribution  -log Transformation 
```

```python
sns.distplot(np.log(l))
```

```python
sns.distplot(np.sqrt(l))
#-square root Transformation 
```

```python
# chi- square distribution 
```

```python
c=np.random.chisquare(df=1,size=10000)
c=c[np.where(abs(c)<10)]
plt.hist(c, bins=1000)
plt.show()
```

```python
# if dof is increasing then it coverting in way of normal distribution 
c=np.random.chisquare(df=5,size=10000)
c=c[np.where(abs(c)<10)]
plt.hist(c, bins=1000)
plt.show()
```

```python
# t - distribution 
```

```python
# t dist with df 1> tail is thicker
t=np.random.standard_t(df=1,size=10000)
t=t[np.where(abs(t)<10)]
plt.hist(t, bins=1000)
plt.show()
```

```python
# t dist with df 10> tail is thinner/smaller
t=np.random.standard_t(df=10,size=10000)
t=t[np.where(abs(t)<10)]
plt.hist(t, bins=1000)
plt.show()
```

```python
# t dist with df 30> tail is more  thinner/smaller and approximately normal dist
t=np.random.standard_t(df=30,size=10000)
t=t[np.where(abs(t)<10)]
plt.hist(t, bins=1000)
plt.show()
```

```python
# F- distribution 
# dfnum is df for numerator , dfden is df for denominator 
f=np.random.f(dfnum=1, dfden=1,size=10000)
f=f[np.where(abs(f)<10)]
plt.hist(f, bins=1000)
plt.show()
```

```python
# dfnum is df for numerator , dfden is df for denominator 
f=np.random.f(dfnum=1, dfden=10,size=10000)
f=f[np.where(abs(f)<10)]
plt.hist(f, bins=1000)
plt.show()
```

```python
# dfnum is df for numerator , dfden is df for denominator 
f=np.random.f(dfnum=3, dfden=10,size=10000)
f=f[np.where(abs(f)<10)]
plt.hist(f, bins=1000)
plt.show()
```
